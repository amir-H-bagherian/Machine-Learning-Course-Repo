importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

importing the test and train datasets

```python
train_dataframe = pd.read_csv('train.csv')
x = train_dataframe.iloc[:, :-1].values
y = train_dataframe.iloc[:, -1].values

test_dataframe = pd.read_csv('test.csv')
x_test = test_dataframe.iloc[:, 0].values
y_test = test_dataframe.iloc[:, -1].values
```

training the simple linear regression model on the training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x, y)
```

Predict the test set results

```python
y_pred = regressor.predict(x_test.reshape(-1, 1))
```

Visualising the training set results

```python
plt.scatter(x, y, color='red')
plt.plot(x, regressor.predict(x), color='blue')
plt.title('some linear reg')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

visualising the test set results

```python
plt.scatter(x_test, y_test, color="red")
plt.plot(x, regressor.predict(x), color="blue")
plt.title('some linear reg test')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```
